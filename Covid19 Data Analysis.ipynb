```python
#importing Covid19 dataset#
#importing covid19_Confirmed_dataset.csv from /Covid19 datastes folder#
covid19_Confirmed_dataset_csv=pd.read_csv("C:\\Users\\Prabhanshu Gautam\\Documents\\Covid19 datasets\covid19_Confirmed_dataset.csv")
covid19_Confirmed_dataset_csv.head(10) #to load csv file or data#
```

```python
#deleting useless data from dataset
covid19_Confirmed_dataset_csv.drop(["Lat","Long"],axis=1,inplace=True)#"method return the copy of covid19_Confirmed_dataset without data_frame.head() 
covid19_Confirmed_dataset_csv.head()                                  #without latitude and longitude [inplace=true]"
```

```python
#aggregating the rows by country 
data_frame_agg=covid19_Confirmed_dataset_csv.groupby("Country/Region").sum() 
data_frame_agg.head(20)
#after the afreggating the dataset is shown by country/region not by indices
```

```python
#find maximum infection rate for all country
countries=list(data_frame_agg.index) #creatinlist of countries having indexes
max_infection=[] #new list for max infectin cases in day
for i in countries: 
    max_infection.append(data_frame_agg.loc[i].diff().max()) #appending all the maximum cases of countries in max_infection
max_infection 
```

```python
data_frame_agg["max_infection_rate"]=max_infection #created new column for maximum ifection rate in country
data_frame_agg.head() 
```

```python
#creating new data frame for with only needed column
infected_data=pd.DataFrame(data_frame_agg["max_infection_rate"])
infected_data.head()
```
